```python
data1 = pd.concat([pop1,
                   sim.get_value('conception_rate')(pop1.index).rename('conception_rate'),
                   sim.get_value('background_mortality_rate')(pop1.index).rename('background_mortality_rate'),
                   #sim.get_value('disability_weight')(pop1.index).rename('disability_weight'),
                   sim.get_value('hemoglobin.exposure_parameters')(pop1.index),
                   sim.get_value('hemoglobin.exposure')(pop1.index).rename('hemoglobin_exposure'),
                   sim.get_value('anemia_levels')(pop1.index).rename('anemia_exposure'),
                  ], axis=1)

    # this concat versus merge function is "risky" usually,
    # but the index is always guarenteed to be sorted here, so it's ok
```

```python
data1['mild_threshold'] = np.where(data1.age<15, 115, np.where(data1.pregnancy_status=='not_pregnant', 120, 110))
data1['moderate_threshold'] = np.where(data1.age<15, 110, np.where(data1.pregnancy_status=='not_pregnant', 110, 100))
data1['severe_threshold'] = np.where(data1.age<15, 80, np.where(data1.pregnancy_status=='not_pregnant', 80, 70))
data1['ali_anemia'] = np.where(data1.hemoglobin_exposure < data1.severe_threshold, 'severe',
                              np.where(data1.hemoglobin_exposure < data1.moderate_threshold, 'moderate',
                                      np.where(data1.hemoglobin_exposure < data1.mild_threshold, 'mild', 'none')))
assert np.all(data1.anemia_exposure == data1.ali_anemia), "Simulated anemia exposures don't match up with expected thresholds"
```

```python
# a loop like this is helpful to collect information about simulants at each timestep
# the InteractiveContext functions (like .run(), etc.) will just give you information
# of simulants at the time when you stop running.


# run for ~ one year so we can see the whole pregnancy transition for some simulants
sim_data = pd.DataFrame()
for step in list(range(0,10)):
    sim.step()
    pop_t = sim.get_population()
    data_t = pd.concat([pop_t,
                   sim.get_value('conception_rate')(pop_t.index).rename('conception_rate'),
                   sim.get_value('background_mortality_rate')(pop_t.index).rename('background_mortality_rate'),
                   #sim.get_value('disability_weight')(pop_t.index).rename('disability_weight'),
                   sim.get_value('hemoglobin.exposure_parameters')(pop_t.index),
                   sim.get_value('hemoglobin.exposure')(pop_t.index).rename('hemoglobin_exposure'),
                   sim.get_value('anemia_levels')(pop_t.index).rename('anemia_exposure'),
                  ], axis=1)
    data_t['step'] = step
    sim_data = pd.concat([sim_data, data_t])
```
